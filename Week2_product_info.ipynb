Validate numeric values

```python
mapping = {
    'nan': 0,
    'three': 3,
    'five': 5,
}
sales_data['quantity'] = sales_data['quantity'].apply(lambda x: mapping.get(x, x))
sales_data.head(10)
```

```python
sales_data['quantity'] = sales_data['quantity'].fillna(0)
sales_data.head(10)
```

```python
sales_data['quantity'].unique()
```

```python
sales_data['quantity'] = pd.to_numeric(sales_data['quantity'],errors='coerce')
```

```python
sales_data['quantity'].dtype
```

```python
#Handling outliers in sales_data by filtering out rows with quantity, unit_price and discount_applied values that are less than 0
sales_data = sales_data[(sales_data["unit_price"] >= 0) & (sales_data["discount_applied"] >= 0) & (sales_data["quantity"] > 0)]
sales_data.head(10)
```

Clean the Data

```python
#Converting the order_date column in sales_data to datetime format for easier analysis and manipulation
sales_data['order_date'] = pd.to_datetime(
    sales_data['order_date'], format='%d-%m-%Y')



```

```python
#Handling outliers in product_info by filtering out rows with base_price values that are less than or equal to 0
product_info['launch_date'] = pd.to_datetime(product_info['launch_date'], format='%d-%m-%y')
product_info = product_info[(product_info['base_price']>0)]
print(product_info['base_price'])
```

Merge the Data

```python
#Merging the cleaned datasets into a single DataFrame for comprehensive analysis,
# using left joins to retain all records from sales_data and matching records from product_info and customer_info
merged_sales_product = pd.merge(sales_data, product_info, on='product_id', how='left')
merged_all = pd.merge(merged_sales_product, customer_info, on='customer_id', how='left')
merged_sales_product.head(10)
```

Feature Engineering

```python
#Converting the quantity, unit_price and discount_applied columns in merged_all to numeric data types for accurate calculations and analysis
cols = ['quantity', 'unit_price', 'discount_applied']
merged_all[cols] = merged_all[cols].apply(pd.to_numeric, errors='coerce')
```

```python
#Adding a new column revenue to merged_all by calculating the total revenue for each transaction based on the quantity, 
# unit_price and discount_applied columns
merged_all['revenue'] = merged_all['quantity'] * merged_all['unit_price'] * (1 - merged_all['discount_applied'])
merged_all.head()
```

```python
#Converting the order_date column in merged_all to datetime format 
# for easier analysis and manipulation
merged_all['order_date'] = pd.to_datetime(merged_all['order_date'], dayfirst=True, errors='coerce')
merged_all
```

```python
#Adding a new column order_week to merged_all by extracting the week number 
# from the order_date column using the dt.isocalendar().week function
merged_all['order_week'] = merged_all['order_date'].dt.isocalendar().week
merged_all
```

```python
#Creating a new column price_band in merged_all by categorizing the unit_price into three bands: 
# Low (0-15), Medium (15-30) and High (30 and above) using the pd.cut() function
bins = [0, 15, 30, float('inf')]
labels = ["Low", "Medium", "High"]
merged_all['price_band'] = pd.cut(merged_all['unit_price'], bins=bins, labels=labels, right=True)
merged_all


```

```python
#Adding a new column days_to_order to merged_all by calculating the number of days between 
# the launch_date and order_date columns using the dt.days attribute
merged_all['days_to_order'] = (merged_all['order_date'] - merged_all['launch_date']).dt.days
merged_all.head(10)
```

```python
#Creating and extracting the email domain from the email column in merged_all and creating a new column email_domain 
# to store this information for further analysis of customer demographics and preferences
merged_all['email_domain'] = merged_all['email'].str.split('@').str[1]
merged_all
```

```python
merged_all['is_late'] = merged_all['delivery_status'] == 'Delayed'
merged_all.tail(10)
```

```python
#Calculating the count of late and on-time deliveries by region 
# using the is_late column and region_y column in merged_all
merged_all.groupby('is_late')['region_y'].value_counts()
```

Create Summary Tables

```python
#Product category performance (revenue, quan4ty, discount)
merged_all.groupby('product_id',as_index= False).agg({'revenue':'sum', 'quantity':'sum', 'discount_applied':'mean'})

```

```python
#Weekly revenue trends by region

merged_all.groupby(['region_y', 'order_week'])['revenue'].sum()
```

```python
#Customer behaviour by loyalty_tier and signup_month
merged_all['signup_date'] = pd.to_datetime(merged_all['signup_date'], errors='coerce')
```

```python
#Mapping the loyalty_tier values to standardize them and handle any inconsistencies in the data, 
# such as different spellings or missing values, 
# to ensure accurate analysis of customer behavior based on loyalty tiers
merged_all['loyalty_tier'] = merged_all['loyalty_tier'].astype(str).str.strip().str.lower()
mapping = {
    'bronze': 'Bronze',
    'silvr': 'Silver',
    'gold': 'Gold',
    'gld': 'Gold',
    'brnze': 'Bronze',
    'Sllver': 'Silver',
    'sllver': 'Silver',
    'nan': 'Unknown'
}
merged_all['loyalty_tier'] = merged_all['loyalty_tier'].replace(mapping).fillna('Unknown')
merged_all['loyalty_tier'] = merged_all['loyalty_tier'].str.capitalize()

```

```python
#Customer behaviour by loyalty_tier and signup_month
merged_all.groupby(['loyalty_tier', merged_all['signup_date'].dt.month], as_index= False)['revenue'].sum().unstack()
```

```python
#Delivery performance by region and price_band
merged_all.groupby(['region_y'])['price_band'].value_counts()
```

```python
#Preferred payment methods by loyalty_4er
merged_all.groupby(['payment_method'])['loyalty_tier'].value_counts()
```

Line plot

```python
#Line plot - weekly revenue trends by region
import matplotlib.pyplot as plt
weekly_revenue = merged_all.groupby(['region_y', 'order_week'])['revenue'].sum().reset_index()
plt.figure(figsize=(13,6))
for region, data in weekly_revenue.groupby('region_y'):
    plt.plot(data['order_week'], data['revenue'], marker='o', label=region)
plt.xlabel('Order Week')
plt.ylabel('Revenue')
plt.title('Weekly Revenue Trends by Region')
plt.legend()
plt.plot()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Prepare the data (using your groupby logic)
pivot_table = merged_all.groupby(['loyalty_tier', merged_all['signup_date'].dt.month])['revenue'].sum().unstack()

# 2. Create the Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu")

# 3. Add labels
plt.title('Total Revenue by Loyalty Tier and Signup Month')
plt.xlabel('Signup Month')
plt.ylabel('Loyalty Tier')
plt.show()
```
